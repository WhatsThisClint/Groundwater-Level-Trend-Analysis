```python
import os
import pandas as pd
import shutil

# Get the directory where the script is located
script_directory = os.path.dirname(os.path.abspath(__file__)) if '__file__' in globals() else os.getcwd()

# Function to find and process the Excel file in folders and subfolders
def process_excel_files_in_folders(folder_path):
    # Loop through all files and subfolders in the current folder
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file == "slope_comparison.xlsx":
                # If the Excel file is found, process it
                excel_file_path = os.path.join(root, file)
                process_excel_file(excel_file_path)

# Function to process the Excel file
def process_excel_file(excel_file_path):
    # Load the Excel file and specify the sheet name
    df = pd.read_excel(excel_file_path, sheet_name="Different")

    # Create the output folder based on the location of the Excel file
    output_folder = os.path.join(os.path.dirname(excel_file_path), "difference graphs")
    os.makedirs(output_folder, exist_ok=True)

    # Loop through the site column and copy corresponding images
    for site_name in df["site"]:
        # Define the source and destination paths for the image
        source_path = os.path.join(os.path.dirname(excel_file_path), "graphs", f"Site_{site_name}.png")
        destination_path = os.path.join(output_folder, f"{site_name}.png")

        # Check if the image file exists before copying
        if os.path.exists(source_path):
            shutil.copy(source_path, destination_path)
            print(f"Copied {site_name}.png to {output_folder}")
        else:
            print(f"Image {site_name}.png not found in {os.path.dirname(excel_file_path)}/graphs")

# Start the search from the script directory
process_excel_files_in_folders(script_directory)

print("Copy process completed.")
```
